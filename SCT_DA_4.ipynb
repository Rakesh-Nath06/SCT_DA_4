### STEP 1: IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
```

### STEP 2: LOAD DATASET

```python
df=pd.read_csv(r"marketing_campaign_performance_10000.csv")
df.head()
```

### STEP 4: CHECK MISSING VALUES

```python
print("Missing Values Per Column:")
print(df.isnull().sum())
```

### STEP 5: CHECK DUPLICATES

```python
print("Duplicate Rows:", df.duplicated().sum())
```

### STEP 6: REMOVE DUPLICATES

```python
df.drop_duplicates(inplace=True)
```

### STEP 7: HANDLE MISSING VALUES

```python
important_cols = ['Channel', 'Impressions', 'Clicks', 'Leads', 'Conversions', 'Cost_USD', 'Revenue_USD']
df.dropna(subset=important_cols, inplace=True)
```

```python
df['Channel'].fillna("Unknown", inplace=True)
```

### STEP 8: REMOVE BLANK CHANNEL NAMES

```python
df = df[df['Channel'].str.strip() != ""]
```

### STEP 9: CHECK INVALID NUMERIC VALUES

```python
df = df[df['Cost_USD'] >= 0] # Negative Spend
```

```python
df = df[df['Revenue_USD'] >= 0] # Negative Revenue
```

```python
df = df[df['Impressions'] > 0]  # Zero Impressions
```

```python
df = df[df['Clicks'] > 0] # Zero Clicks
```

### STEP 10: CONVERT DATE COLUMNS

```python
df['StartDate'] = pd.to_datetime(df['StartDate'])
df['EndDate'] = pd.to_datetime(df['EndDate'])
```

### STEP 11: CREATE CAMPAIGN DURATION

```python
df['Campaign_Duration'] = (df['EndDate'] - df['StartDate']).dt.days
```

### STEP 12: CREATE KPI METRICS

```python
# CTR (Click through Rate)
```

```python
df['CTR'] = (df['Clicks'] / df['Impressions']) * 100
```

```python
# Conversion Rate
```

```python
df['Conversion_Rate'] = (df['Conversions'] / df['Clicks']) * 100
```

```python
# Lead Rate
```

```python
df['Lead_Rate'] = (df['Leads'] / df['Clicks']) * 100
```

```python
# CPC (Cost Per Click)
```

```python
df['CPC'] = df['Cost_USD'] / df['Clicks']
```

```python
# CPA (Cost per Acquision)
```

```python
df['CPA'] = df['Cost_USD'] / df['Conversions']
```

```python
# ROI (Return on Investment) 
```

```python
df['ROI'] = ((df['Revenue_USD'] - df['Cost_USD']) / df['Cost_USD']) * 100
```

### STEP 13: HANDLE DIVISION ERRORS

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
```

### STEP 14: SAVE CLEANED DATASET

```python
df.to_csv("cleaned_marketing_campaign_data.csv", index=False)
```

### STEP 15: CHANNEL PERFORMANCE SUMMARY

```python
channel_summary = df.groupby('Channel').agg({
    'Cost_USD': 'sum',
    'Revenue_USD': 'sum',
    'Conversions': 'sum',
    'ROI': 'mean',
    'CPA': 'mean',
    'Conversion_Rate': 'mean'
}).reset_index()
```

```python
print(channel_summary)
```

### STEP 16: BEST REVENUE CHANNEL

```python
best_revenue = channel_summary.loc[channel_summary['Revenue_USD'].idxmax()]
print("Highest Revenue Channel:")
print(best_revenue)
```

### STEP 17: BEST ROI CHANNEL

```python
best_roi = channel_summary.loc[channel_summary['ROI'].idxmax()]
print("Highest ROI Channel:")
print(best_roi)
```

### STEP 18: LOWEST CPA CHANNEL

```python
lowest_cpa = channel_summary.loc[channel_summary['CPA'].idxmin()]
print("Lowest CPA Channel:")
print(lowest_cpa)
```

### STEP 19: WORST CHANNEL (Budget Waste)

```python
worst_roi = channel_summary.loc[channel_summary['ROI'].idxmin()]
print("Lowest ROI / Budget Waste Channel:")
print(worst_roi)
```

### STEP 20: CALCULATE FUNNEL TOTALS

```python
funnel = {
    'Impressions': df['Impressions'].sum(),
    'Clicks': df['Clicks'].sum(),
    'Leads': df['Leads'].sum(),
    'Conversions': df['Conversions'].sum()
}

print("Funnel Totals:")
print(funnel)
```

### STEP 21: FUNNEL DROP-OFF RATES

```python
click_drop = (funnel['Clicks'] / funnel['Impressions']) * 100  # Click-through from impressions
```

```python
lead_drop = (funnel['Leads'] / funnel['Clicks']) * 100 # Leads from clicks
```

```python
conversion_drop = (funnel['Conversions'] / funnel['Leads']) * 100 # Conversions from leads
```

```python
print("Click Rate from Impressions:", click_drop)
print("Lead Rate from Clicks:", lead_drop)
print("Conversion Rate from Leads:", conversion_drop)
```

### STEP 22: INTERPRETATION LOGIC

```python
if click_drop < 5:                                         # Example
    print("Poor ad creatives or weak targeting.")
    
if lead_drop < 20:
    print("Landing page optimization needed.")
    
if conversion_drop < 30:
    print("Sales or nurturing process needs improvement.")
```

### STEP 23: MONTHLY PERFORMANCE ANALYSIS

```python
df['Month'] = df['StartDate'].dt.to_period('M')

monthly_summary = df.groupby('Month').agg({
    'Cost_USD':'sum',
    'Revenue_USD':'sum',
    'Conversions':'sum',
    'ROI':'mean'
}).reset_index()

print(monthly_summary)
```
